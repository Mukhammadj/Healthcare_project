## Data Visualization and Constructing Hypothesis 

```python
#Let's find out correlation between features
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df.corr())
plt.title('Correlation between features',color = 'purple',fontsize = 20)
plt.savefig('corr_overall.png', dpi=300)
plt.show()
```

```python
#Correlation of features to the terget
corr = df[df.columns[1:]].corr()['Persistency_Flag'][:]
corr_sorted = corr.sort_values(ascending=False)

#Let's see top 10 most correlated features 
corr_sorted[:10]
```

```python
# We keep columns(>25) in which scores are higher than 25. And 'Gender', 'Ethnicity', 'Age_Bucket', 'Region'

columns = ['Persistency_Flag', 'Gender', 'Ethnicity', 'Age_Bucket', 'Region', 
          'Dexa_During_Rx', 'Dexa_Freq_During_Rx', 'Comorb_Long_Term_Current_Drug_Therapy',
         'Comorb_Encounter_For_Screening_For_Malignant_Neoplasms', 'Comorb_Encounter_For_Immunization']
df_nes = df[columns]
df_nes.head()
```

```python
f, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df_nes.corr(), annot=True)
plt.title('Correlation between features',color = 'purple',fontsize = 20)
plt.savefig('corr_nes.png', dpi=300)
plt.show()
```

```python
# Checking the ratio of the target variable (Persistency_Flag)

plt.bar(['0', '1'], df_nes['Persistency_Flag'].value_counts(ascending=True))
plt.show()
```

```python
 # Gender wise analysis
    
pd.crosstab(df0['Gender'], df_nes['Persistency_Flag']).plot(kind='bar', figsize=(15, 6))
plt.title('Frequency for Gender')
plt.xlabel('Gender')
plt.legend(['Not Persistent', 'Persistent'])
plt.show()
```

```python
df_nes.groupby([df0["Age_Bucket"].sort_values(ascending=False), df0["Gender"]])["Persistency_Flag",].mean()
```

```python

pd.crosstab(df['Count_Of_Risks'], df_nes['Persistency_Flag']).plot(kind='bar', figsize=(15, 6))
plt.legend(['Not Persistent', 'Persistent'])
plt.show()
```

```python
#By region
pd.crosstab(df0['Region'], df_nes['Persistency_Flag']).plot(kind='bar', figsize=(15, 6))
plt.legend(['Not Persistent', 'Persistent'])
plt.show()
```

<strong> The highest average persistency occured in '<55' aged group for both genders </strong>
